```python
import pandas as pd
print("=== Carregando dados ===")

# 1. Carregando os dados (Usando caminhos relativos corretos)
transactions = pd.read_parquet("../data/raw/transactions.parquet")
merchants = pd.read_parquet("../data/raw/merchants.parquet")

# 2. Cruzando as tabelas (Fato + Dimensão)
df = transactions.merge(merchants, on="merchant_id", how="left")

# --- MÉTRICAS GERAIS ---
print("=== INDICADORES GERAIS ===")
tpv_total = df["amount"].sum()
ticket_medio = df["amount"].mean()

# Formatando os números para a moeda local (R$)
print(f"TPV Total: R$ {tpv_total:,.2f}")
print(f"Ticket Médio Geral: R$ {ticket_medio:,.2f}")
print("\n" + "="*30 + "\n")

# --- ANÁLISES AGRUPADAS ---
print("=== TICKET MÉDIO POR SEGMENTO ===")
# Salvando como DataFrame para ficar com uma visualização bonita no Notebook
ticket_segmento = df.groupby("segment")["amount"].mean().reset_index().rename(columns={"amount": "ticket_medio"})
display(ticket_segmento)

print("\n=== TPV POR ESTADO ===")
tpv_estado = df.groupby("state")["amount"].sum().reset_index().rename(columns={"amount": "tpv"}).sort_values(by="tpv", ascending=False)
display(tpv_estado)

print("\n=== VOLUMETRIA POR BANDEIRA ===")
# Cruzando quantidade de vendas e TPV total por bandeira na mesma tabela
analise_bandeira = df.groupby("card_brand")["amount"].agg(qtd_transacoes="count", tpv_total="sum").reset_index()
display(analise_bandeira)
```
